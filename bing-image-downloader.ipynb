```python
from bing_image_downloader import downloader

query = 'rotten food'
downloader.download(query, limit=200, output_dir='train_set', adult_filter_off=True)
```

```python
import os

file_path = os.path.join("train_set/rotten food")

for i, filename in enumerate(os.listdir(file_path)):
    ext = os.path.splitext(filename)[-1]
    new_name = f"{query.replace(' ', '_')}_{i+1:04d}{ext}"
    old_path = os.path.join(file_path, filename)
    new_path = os.path.join(file_path, new_name)
    os.rename(old_path, new_path)

print("Penamaan selesai")
```
